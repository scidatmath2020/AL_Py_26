## Matrices simétricas

Una matriz $M$ se dice **simétrica** si

$$
M = M^{t},
$$

donde $M^{t}$ denota la **matriz transpuesta** de $M$.

Notemos lo siguiente. Si $M$ es una matriz de dimensiones $m\times n$, entonces su transpuesta $M^t$ tiene dimensiones $n\times m$.  

Por lo tanto, si

$$
M = M^t,
$$

debe cumplirse que

$$
m = n.
$$

Es decir, $M$ debe ser una **matriz cuadrada** (tiene la misma cantidad de filas que de columnas).

---

## El espacio de matrices cuadradas

Sabemos que el conjunto de matrices reales de tamaño $n\times n$

$$
M_{n\times n}(\mathbb{R})
$$

forma un **espacio vectorial**.

Dentro de este espacio, el conjunto de **matrices simétricas** forma un **subespacio vectorial**.

Para ver esto, recordemos dos propiedades importantes de la transposición:

$$
(A+B)^t = A^t + B^t
$$

y

$$
(aA)^t = aA^t,
$$

donde $a$ es un número real.

---

## Cerradura bajo suma

Supongamos que $A$ y $B$ son matrices **simétricas**.  

Esto significa que

$$
A^t = A, \qquad B^t = B.
$$

Entonces,

$$
(A+B)^t = A^t + B^t = A + B.
$$

Por lo tanto, $A+B$ también es simétrica.

---

## Cerradura bajo multiplicación por escalares

Sea $a\in\mathbb{R}$ y sea $A$ una matriz simétrica.

Entonces,

$$
(aA)^t = aA^t = aA.
$$

Por lo tanto, $aA$ también es simétrica.

---

## Conclusión

El conjunto de matrices simétricas:

- contiene a la matriz cero,
- es cerrado bajo suma,
- es cerrado bajo multiplicación por escalares.

Por lo tanto, el conjunto de matrices simétricas es un **subespacio vectorial de**

$$
M_{n\times n}(\mathbb{R}).
$$

# Span de vectores

Tomemos ahora los vectores

$$
(1,2), \qquad (-3,0)
$$

en $\mathbb{R}^2$.

El **span** de estos vectores consiste en **todas las combinaciones lineales** de ellos:

$$
a(1,2) + b(-3,0), \qquad a,b\in\mathbb{R}.
$$

Calculemos esta combinación:

$$
a(1,2) = (a,2a)
$$

y

$$
b(-3,0) = (-3b,0).
$$

Sumando,

$$
(a,2a) + (-3b,0) = (a-3b,\,2a).
$$

Por lo tanto, el span está formado por todos los vectores de la forma

$$
(a-3b,\,2a), \qquad a,b\in\mathbb{R}.
$$
