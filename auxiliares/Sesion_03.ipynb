Los elementos de $W$ son elementos de $\mathbb{R}^3$ (recordemos que $\mathbb{R}^3$ son ternas ordenadas de números reales).

Un elemento $(x,y,z)$ está en $W$ si

$$
2x + 3y + 2026z = 0.
$$

Por ejemplo, el vector $(0,0,0)$ está en $W$, pues

$$
2\cdot 0 + 3\cdot 0 + 2026\cdot 0 = 0.
$$

En cambio, el vector $(1,1,1)$ **no vive en $W$**. Hay que cobrarle renta por canijo, porque

$$
2\cdot 1 + 3\cdot 1 + 2026\cdot 1 = 2031 \neq 0.
$$

Para vivir en $W$, la operación anterior debió dar $0$.

---

## El conjunto $W$

Consideremos el conjunto

$$
W = \{(x,y,z) \in \mathbb{R}^3 \mid 2x + 3y + 2026z = 0\}.
$$

Queremos entender por qué $W$ es un **subespacio vectorial de $\mathbb{R}^3$**.

---

## 1. El vector cero pertenece a $W$

El vector $(0,0,0)$ debe vivir en $W$. En efecto,

$$
2\cdot0 + 3\cdot0 + 2026\cdot0 = 0.
$$

Por lo tanto,

$$
(0,0,0) \in W.
$$

---

## 2. Cerradura bajo suma

Tomemos dos elementos de $W$, digamos

$$
(a,b,c), \quad (p,q,r).
$$

Esto significa que

$$
2a + 3b + 2026c = 0
$$

y

$$
2p + 3q + 2026r = 0.
$$

La suma de estos vectores es

$$
(a,b,c) + (p,q,r) = (a+p,\, b+q,\, c+r).
$$

Veamos si este vector también vive en $W$.

Calculamos

$$
2(a+p) + 3(b+q) + 2026(c+r).
$$

Desarrollando,

$$
\begin{aligned}
2(a+p)+3(b+q)+2026(c+r)
&=2a+2p+3b+3q+2026c+2026r \\
&=(2a+3b+2026c)+(2p+3q+2026r) \\
&=0+0 \\
&=0.
\end{aligned}
$$

Por lo tanto,

$$
(a+p,\, b+q,\, c+r) \in W.
$$

Esto muestra que **$W$ es cerrado bajo suma**.

---

## 3. Cerradura bajo multiplicación por escalares

Ahora tomemos un número real $t$ y un vector $(a,b,c) \in W$.

Esto significa que

$$
2a + 3b + 2026c = 0.
$$

Multiplicamos el vector por el escalar:

$$
t(a,b,c) = (ta,\, tb,\, tc).
$$

Verifiquemos si este vector vive en $W$.

Calculamos

$$
2(ta) + 3(tb) + 2026(tc).
$$

Factorizando $t$,

$$
2ta + 3tb + 2026tc = t(2a + 3b + 2026c).
$$

Pero como $(a,b,c)\in W$,

$$
2a + 3b + 2026c = 0.
$$

Entonces,

$$
t(2a+3b+2026c) = t\cdot 0 = 0.
$$

Por lo tanto,

$$
(ta,tb,tc) \in W.
$$

Esto muestra que **$W$ es cerrado bajo multiplicación por escalares**.

---

## Conclusión

Hemos demostrado que:

- $(0,0,0) \in W$
- $W$ es cerrado bajo suma
- $W$ es cerrado bajo multiplicación por escalares

Por lo tanto,

$$
W
$$

es un **subespacio vectorial de $\mathbb{R}^3$**.
