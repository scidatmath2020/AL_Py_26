# Introducción 

**Una introducción a los fundamentos del álgebra lineal a través del piano**

El álgebra lineal no es, en esencia, un conjunto de reglas abstractas o un catálogo de operaciones matriciales. Es el estudio de cómo se combinan elementos básicos —vectores— para generar estructuras más complejas, y cómo esas combinaciones revelan patrones ocultos en los datos.

Para comprenderlo de manera intuitiva y profunda, recurriremos a un instrumento que ha sido durante siglos un modelo perfecto de combinación armónica: el piano. Sus teclas no son meras notas aisladas, sino generadores de un espacio sonoro donde cada presión, cada superposición, cada escala define posibilidades infinitas a partir de un conjunto finito de bases.

Comencemos por lo elemental.

## La tecla individual: el vector

Presione una sola tecla —digamos, el Do central—. El sonido es puro, unidireccional: tiene una frecuencia definida, una amplitud (intensidad) y una posición fija en el teclado.  

Esto es precisamente un **vector**: una entidad con magnitud y dirección. En el contexto de los datos, una variable (columna de un dataset) es análoga a una tecla: representa una dimensión específica —edad, temperatura, precio, frecuencia de uso— con un valor que apunta en una "dirección" particular del espacio de observaciones.

Una tecla por sí sola no constituye una pieza musical completa, del mismo modo que una sola variable rara vez captura la complejidad de un fenómeno. Pero es el ladrillo fundamental: sin vectores base no hay estructura posible.



## La superposición controlada: la combinación lineal

Ahora presiona simultáneamente Do, Mi y Sol —un acorde mayor—. El resultado no es la suma caótica de tres sonidos independientes, sino una entidad nueva, coherente, con su propia identidad armónica.  

Así, has realizado una **combinación lineal**: has tomado vectores (teclas) y los has multiplicado por coeficientes (la fuerza o "volumen" con que presiona cada una) antes de sumarlos.  

En notación:  
**$$c_1v_1 + c_2v_2 + c_3v_3$$**  

Donde los $c_i$ son escalares reales (positivos o negativos, mayores o menores que 1), y los $v_i$ son los vectores base.  

En datos, cada fila de un dataset (una observación) es exactamente eso: una combinación lineal de las variables (columnas), ponderada por los valores medidos. Un cliente no es "edad + ingresos + clics"; es una mezcla específica: 35 años × 1 + $45,000 × 1 + 120 clics × 0.8, etc.  

La potencia radica en que, ajustando los coeficientes, puede recrear cualquier punto dentro del espacio generado por esas bases.

## El alcance del teclado: el espacio generado (span)

Con solo un puñado de teclas —las doce de una octava cromática— el piano puede producir prácticamente toda la música tonal occidental. Ese conjunto finito genera un espacio vasto de combinaciones posibles.  

En álgebra lineal llamamos **span** al conjunto de todas las combinaciones lineales posibles de un conjunto de vectores. Si los vectores son linealmente independientes (ninguno es combinación de los otros, como teclas que no están en la misma "afinación" redundante), el span adquiere la dimensión igual al número de vectores.

En análisis de datos, la elección de variables iniciales define el span de su modelo:  
- Variables independientes → span amplio, capacidad explicativa alta.  
- Variables altamente correlacionadas → span reducido, redundancia que oculta señal bajo ruido.

El arte consiste en seleccionar —o transformar— las bases para que el span capture lo esencial sin desperdiciar dimensiones.

## Hacia adelante: el viaje del diplomado

Este capítulo introductori establece las bases intuitivas: vector, combinación lineal, espacio generado.  

En los siguientes formalizaremos estos conceptos, además de aprender:  
- Cómo cambiar de base (transiciones entre tonalidades, matrices de cambio de base).  
- Direcciones privilegiadas en el espacio (eigenvectores y eigenvalores).  
- Descomposición de estructuras complejas (SVD).  
- Reducción de dimensionalidad preservando armonía (PCA).  

Cada paso se construye sobre la misma lógica: el piano no cambia de naturaleza al pasar de una escala a otra; solo se reinterpreta. Lo mismo ocurre con los datos.

El álgebra lineal, vista así, deja de ser un obstáculo técnico para convertirse en una herramienta de composición: permite descubrir la música subyacente en el ruido de los números.
